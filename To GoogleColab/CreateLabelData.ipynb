```python
def xml_to_df(path):
    xml_list = []
    #print(path + '*.xml')
    for xml_file in glob.glob(path + '*.xml'):
        #print(xml_file)
        tree = ET.parse(xml_file)
        root = tree.getroot()
        for member in root.findall('object'):
            value = (root.find('filename').text,
                     int(root.find('size')[0].text),
                     int(root.find('size')[1].text),
                     member[0].text,
                     int(member[4][0].text),
                     int(member[4][1].text),
                     int(member[4][2].text),
                     int(member[4][3].text)
                     )
            xml_list.append(value)
    column_name = ['filename', 'width', 'height', 'class', 'xmin', 'ymin', 'xmax', 'ymax']
    xml_df = pd.DataFrame(xml_list, columns=column_name)
    return xml_df
```

```python
image_path = g_drive + 'annotated/'
xml_df = xml_to_df(image_path)
```

```python
xml_df.head()
```

```python
fileName = np.sort(xml_df["filename"].unique())
```

```python
fileName = np.sort(xml_df["filename"].unique())
textAry = []
for fn in fileName:
  textLine = 'images/' + fn
  tempDf = xml_df[xml_df['filename'] == fn]
  for index, d in tempDf.iterrows():
    textLine =  ("%s %d,%d,%d,%d,0" % (textLine, d.xmin, d.xmax, d.ymin, d.ymax))
    #print(textLine)
  textAry.append(textLine)
```

```python
xml_df.to_csv(g_drive + 'labels.csv', index=None)
print('Successfully converted xml to csv.')
```

```python
with open(g_drive + "my_train.txt", 'wt') as f:
  for ele in textAry:
    f.write(ele+'\n')
```
